```python
import plotly.graph_objects as go

for z in range(51):
    try:
        indice_gestor = df.loc[(df['Quem será avaliado?'] == f'Funcionário{z}') & (df['Qual seu nome?'] != '-')].index[0]
    except:
        pass

    try:
        indice_auto = df.loc[df['Qual seu nome?'] == f'Funcionário{z}'].index[0]
    except:
        pass

    try:
        indice_anonimo = df.loc[(df['Quem será avaliado?'] == f'Funcionário{z}') & (df['Qual seu nome?'] == '-')].index[0]
    except:
        pass
    
    a = -3
    b = 2  
    c = -1  

    def radar(a, b):
        fig = go.Figure()

        fig.add_trace(go.Scatterpolar(
            r=list(df.iloc[indice_gestor, a:b]),
            theta=list(df.columns)[a:b],
            fill='toself',
            name='Avaliação do Gestor'
        ))

        fig.add_trace(go.Scatterpolar(
            r=list(df.iloc[indice_auto, a:b]),
            theta=list(df.columns)[a:b],
            fill='toself',
            name='Auto Avaliação'
        ))

        fig.add_trace(go.Scatterpolar(
            r=list(df.iloc[indice_anonimo, a:b]),
            theta=list(df.columns)[a:b],
            fill='toself',
            name='Avaliação Anônima (média)'
        ))

        fig.update_layout(
        polar=dict(
            radialaxis=dict(
            visible=True,
            range=[0, 10]
            )),
        showlegend=True, legend=dict(y=1.3)
        )

        fig.write_image(f"images/{main_categories[c]}_funcionario{z}.png")

    

    for i in range(len(main_categories)):
        a += 5
        b += 5
        c += 1
        radar(a, b)
        
```

```python
from fpdf import FPDF

for i in range(51):
    class PDF(FPDF):
        def header(self):
            # Logo
            self.image('fiemt.png', 10, 8, 40)
            # Arial bold 15
            self.set_font('Arial', 'B', 15)
            # Move to the right
            self.cell(80)
            # Title
            self.cell(30, 10, 'Relatório Percepção de Valores', 0, 0, 'C')
            # Line break
            self.ln(20)

        # Page footer
        def footer(self):
            # Position at 1.5 cm from bottom
            self.set_y(-15)
            # Arial italic 8
            self.set_font('Arial', '', 8)
            # Page number
            self.cell(0, 10, 'Page ' + str(self.page_no()) + '/{nb}', 0, 0, 'C')

    # Instantiation of inherited class
    pdf = PDF()
    pdf.alias_nb_pages()
    pdf.add_page(orientation='portrait')

    pdf.set_font('Times', 'B', 16)
    pdf.cell(0, 20, f'Colaborador: Funcionário{i}', 0, 2)

    # Introdução
    pdf.set_font('Times', '', 14)
    intro = open('textos/introducao.txt', 'r')
    x = intro.readline()
    pdf.multi_cell(w=190, h=5, txt= x, border = 0, align= 'J')

    # Texto 1
    pdf.set_font('Times', 'B', 16)
    pdf.cell(0, 20, main_categories[0], 0, 2, align='C')

    pdf.set_font('Times', '', 14)
    integridade = open('textos/integridade.txt', 'r')
    y = integridade.readline()
    pdf.multi_cell(w=190, h=5, txt= y, border = 0, align= 'J')
    pdf.image(f'images/Integridade_funcionario{i}.png', type='png', w=175*1.1, h=125*1.1)

    # Texto 2
    pdf.set_font('Times', 'B', 16)
    pdf.cell(0, 20, main_categories[1], 0, 2, align='C')

    pdf.set_font('Times', '', 14)
    integridade = open('textos/valorizacao.txt', 'r')
    y = integridade.readline()
    pdf.multi_cell(w=190, h=5, txt= y, border = 0, align= 'J')
    pdf.image(f'images/Valorização das Pessoas_funcionario{i}.png', type='png', w=175*1.1, h=125*1.1)

    # Texto 3
    pdf.set_font('Times', 'B', 16)
    pdf.cell(0, 20, main_categories[2], 0, 2, align='C')

    pdf.set_font('Times', '', 14)
    integridade = open('textos/uniao.txt', 'r')
    y = integridade.readline()
    pdf.multi_cell(w=190, h=5, txt= y, border = 0, align= 'J')
    pdf.image(f'images/União_funcionario{i}.png', type='png', w=175*1.1, h=125*1.1)

    # Texto 4
    pdf.set_font('Times', 'B', 16)
    pdf.cell(0, 20, main_categories[3], 0, 2, align='C')

    pdf.set_font('Times', '', 14)
    integridade = open('textos/confianca.txt', 'r')
    y = integridade.readline()
    pdf.multi_cell(w=190, h=5, txt= y, border = 0, align= 'J')
    pdf.image(f'images/Confiança nas Relações_funcionario{i}.png', type='png', w=175*1.1, h=125*1.1)

    # Texto 5
    pdf.set_font('Times', 'B', 16)
    pdf.cell(0, 20, main_categories[4], 0, 2, align='C')

    pdf.set_font('Times', '', 14)
    integridade = open('textos/simplicidade.txt', 'r')
    y = integridade.readline()
    pdf.multi_cell(w=190, h=5, txt= y, border = 0, align= 'J')
    pdf.image(f'images/Simplicidade_funcionario{i}.png', type='png', w=175*1.1, h=125*1.1)

    # Texto 6
    pdf.set_font('Times', 'B', 16)
    pdf.cell(0, 20, main_categories[5], 0, 2, align='C')

    pdf.set_font('Times', '', 14)
    integridade = open('textos/mentalidade.txt', 'r')
    y = integridade.readline()
    pdf.multi_cell(w=190, h=5, txt= y, border = 0, align= 'J')
    pdf.image(f'images/Mentalidade de Crescimento_funcionario{i}.png', type='png', w=175*1.1, h=125*1.1)

    # Texto 7
    pdf.set_font('Times', 'B', 16)
    pdf.cell(0, 20, main_categories[6], 0, 2, align='C')

    pdf.set_font('Times', '', 14)
    integridade = open('textos/sucesso.txt', 'r')
    y = integridade.readline()
    pdf.multi_cell(w=190, h=5, txt= y, border = 0, align= 'J')
    pdf.image(f'images/Sucesso do Cliente_funcionario{i}.png', type='png', w=175*1.1, h=125*1.1)


    pdf.output(f'relatorios/funcionario{i}.pdf', 'F')
```
